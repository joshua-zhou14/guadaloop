```python
import pandas as pd
import matplotlib.pyplot as plt
from tabulate import tabulate
from IPython.display import display

data = pd.read_csv('data.csv')
```

```python
tmilekey = ['tmiles_2017', 'tmiles_2018', 'tmiles_2019', 'tmiles_2020', 'tmiles_2021', 'tmiles_2022', 'tmiles_2023',
            'tmiles_2025', 'tmiles_2030', 'tmiles_2035', 'tmiles_2040', 'tmiles_2045', 'tmiles_2050']
tmile = []
for i in range(0, len(tmilekey)):
    tmile.append(0)

tonkey = ['tons_2017', 'tons_2018', 'tons_2019', 'tons_2020', 'tons_2021', 'tons_2022', 'tons_2023',
            'tons_2025', 'tons_2030', 'tons_2035', 'tons_2040', 'tons_2045', 'tons_2050']
tons = []
commodityout = []
for i in range(0, len(tonkey)):
    tons.append(0)
    commodityout.append({})

commoditydict = {}

with open('commodity.txt') as f:
    lines = f.readlines()
    for line in lines:
        temp = line.split(',')
        commoditydict[int(temp[0])] = temp[1].strip()
        for i in range(0, len(commodityout)):
            commodityout[i][temp[1].strip()] = 0


dbandlowkey = {1: 1, 2: 100, 3:250, 4:500, 5:750, 6:1000, 7:1500, 8:2001}
dbandhighkey = {1: 99, 2: 249, 3:499, 4:749, 5:999, 6:1499, 7:2000, 8:2001}
distbandlow = 0
distbandhigh = 0
timesensitivekey = {'Live animals/fish', 'Meat/seafood', 'Alcoholic beverages', 'Tobacco prods.', 'Pharmaceuticals', 'Fertilizers', 'Chemical prods.', 
                    'Newsprint/paper', 'Paper articles', 'Printed prods.', 'Textiles/leather'}

for x in data.index:
    if x == 0:
        continue
    for i in range(0, len(tmilekey)):
        tmile[i] += data.loc[x, tmilekey[i]]
        tons[i] += data.loc[x, tonkey[i]]
        commodityout[i][commoditydict[data.loc[x, "sctg2"]]] += data.loc[x, tonkey[i]]
    distbandlow += dbandlowkey[data.loc[x,"dist_band"]]
    distbandhigh += dbandhighkey[data.loc[x,"dist_band"]]

df = pd.DataFrame({'Commodity': commoditydict.values()})
# dfnew = df

totalpctg = {'2017': 0, '2018': 0, '2019':0, '2020':0, '2021':0, '2022':0, '2023':0, '2025': 0, '2030': 0, 
             '2035': 0, '2040': 0, '2045': 0, '2050': 0}

for i in range(0, len(commodityout)):
    for key in commodityout[i]:
        commodityout[i][key] = round(commodityout[i][key]/tons[i]*100,3)
    df.insert(i+1, tonkey[i][5:],  commodityout[i].values())
#tons_2023 - tons[6]
for x in df.index:
    if(df.loc[x,'Commodity'] not in timesensitivekey):
        df = df.drop(x)
    else:
        for key in totalpctg.keys():
            totalpctg[key] = round(totalpctg[key] + df.loc[x, key], 3)
df.sort_values(by='2017', ascending=False, inplace=True)
temp = ['Total']
for val in totalpctg.values():
    temp.append(val)
df.loc[11] = temp
display(df)
print(totalpctg)
```

```python
#calculating operating time assuming 1 pod
average_human = 70
pod_low = 23
pod_high = 40
ton_to_kilo = 907.18474
mph = 670
pod_low_weight = (pod_low*average_human)/ton_to_kilo
pod_high_weight = (pod_high*average_human)/ton_to_kilo
print(str(pod_low_weight) + " " + str(pod_high_weight))

commoditydict = {}

with open('commodity.txt') as f:
    lines = f.readlines()
    for line in lines:
        temp = line.split(',')
        commoditydict[int(temp[0])] = temp[1].strip()

timesensitivekey = {'Live animals/fish', 'Meat/seafood', 'Alcoholic beverages', 'Tobacco prods.', 'Pharmaceuticals', 'Fertilizers', 'Chemical prods.', 
                    'Newsprint/paper', 'Paper articles', 'Printed prods.', 'Textiles/leather'}
for x in data.index:
    if x == 0:
        continue
    if(commoditydict[data.loc[x,'sctg2']] not in timesensitivekey):
        data = data.drop(x)

datalow = data
datahigh = data
tonkey = ['tons_2017', 'tons_2018', 'tons_2019', 'tons_2020', 'tons_2021', 'tons_2022', 'tons_2023',
            'tons_2025', 'tons_2030', 'tons_2035', 'tons_2040', 'tons_2045', 'tons_2050']
valkey = ['value_2017', 'value_2018', 'value_2019', 'value_2020', 'value_2021', 'value_2022', 'value_2023',
            'value_2025', 'value_2030', 'value_2035', 'value_2040', 'value_2045', 'value_2050']

dbandlowkey = {1: 1, 2: 100, 3:250, 4:500, 5:750, 6:1000, 7:1500, 8:2001}
dbandhighkey = {1: 99, 2: 249, 3:499, 4:749, 5:999, 6:1499, 7:2000, 8:2001}

display(datalow['dist_band'])
datalow.to_csv('data1.csv', encoding='utf-8', index=False)
for x in datalow.index:
    if x == 0: 
        continue
    for key in tonkey:
        datalow.at[x,key] = round(datalow.loc[x,key]*1000/pod_low_weight)
        datahigh.at[x,key] = round(datahigh.loc[x,key]*1000/pod_high_weight)
    #time per trip in hours
    if(datalow.loc[x,"dist_band"] < 1):
        continue
    datalow.at[x, 'dist_band'] = dbandlowkey[round(datalow.loc[x,"dist_band"])]/mph
    if(datahigh.loc[x,"dist_band"] < 1):
        continue
    datahigh.at[x, 'dist_band'] = dbandhighkey[round(datahigh.loc[x,"dist_band"])]/mph
    #find total time per route
    for i in range(0, len(valkey)):
        datalow.at[x,valkey[i]] = round(datalow.loc[x, tonkey[i]]*datalow.loc[x, 'dist_band'])
        datahigh.at[x,valkey[i]] = round(datahigh.loc[x, tonkey[i]]*datahigh.loc[x, 'dist_band'])

dflow = pd.DataFrame({})
dflowtotal = pd.DataFrame({})
dfhigh = pd.DataFrame({})
dfhightotal = pd.DataFrame({})
count = 0
# dflow.insert(count, "Time per Trip", datalow['dist_band'])
# dfhigh.insert(count, "Time per Trip", datahigh['dist_band'])
# count += 1
for key in tonkey:
    temp = "Trips " + key[5:]
    dflow.insert(count, temp, datalow[key])
    dfhigh.insert(count, temp, datahigh[key])
    count += 1
count = 0
for key in valkey:
    temp = "Total Time " + key[6:]
    dflowtotal.insert(count, temp, datalow[key])
    dfhightotal.insert(count, temp, datahigh[key])
    count += 1


dflow.sort_values(by='Trips 2017', ascending=False, inplace=True)
dflowtotal.sort_values(by='Total Time 2017', ascending=False, inplace=True)
dfhigh.sort_values(by='Trips 2017', ascending=False, inplace=True)
dfhightotal.sort_values(by='Total Time 2017', ascending=False, inplace=True)
dflowtotal = dflowtotal.drop(2818)
dfhightotal = dfhightotal.drop(2818)

days = 353
hours = 24
capacity = 0.3

count = 0
dflownum = pd.DataFrame({})
dfhighnum = pd.DataFrame({})
for key in tonkey:
    temp = "Pod Number " + key[5:]
    temp2 = "Total Time " + key[5:]
    dflownum.insert(count, temp, dflowtotal[temp2])
    dfhighnum.insert(count, temp, dfhightotal[temp2])
    count+=1

for x in dflownum.index:
    for key in tonkey:
        temp = "Pod Number " + key[5:]
        dflownum.loc[x, temp] = round(dflownum.loc[x, temp]/days/hours*(1/capacity), 3)
        dfhighnum.loc[x, temp] = round(dfhighnum.loc[x, temp]/days/hours*(1/capacity), 3)

arr = ['2017', '2018', '2019', '2020', '2021', '2022']
totalarr = arr.copy()
numarr = arr.copy()

for i in range(0, len(arr)):
    arr[i] = "Trips " + arr[i]
    totalarr[i] = "Total Time " + totalarr[i]
    numarr[i] = "Pod Number " + numarr[i]

dflownum.sort_values(by='Pod Number 2017', ascending=False, inplace=True)
dfhighnum.sort_values(by='Pod Number 2017', ascending=False, inplace=True)

dflow.drop(arr, axis=1, inplace=True)
dfhigh.drop(arr, axis=1, inplace=True)
dflowtotal.drop(totalarr, axis=1, inplace=True)
dfhightotal.drop(totalarr, axis=1, inplace=True)

display(dflow.head(10))
display(dflowtotal.head(10))
display(dfhigh.head(10))
display(dfhightotal.head(10))

dflownum.drop(numarr, axis=1, inplace=True)
dfhighnum.drop(numarr, axis=1, inplace=True)
display(dflownum.head(10))
display(dfhighnum.head(10))
```
